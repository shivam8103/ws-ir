## Suppose an Information Retrieval (IR) system contains a document collection of 1500
documents, numbered ``d1 to d1500``. A user query returns 30 relevant documents, known in
advance and listed below:

### Relevant Documents:
``{d2, d10, d20, d35, d48, d60, d75, d80, d120, d125, d150, d175, d200, d225, d230, d300, d325, d400, d405, d450, d480, d500, d550, d600, d700, d720, d800, d850, d900, d1000}``

### Two IR systems are used to retrieve only the top 10 documents, in the following ranked order:
####  System A: ``d10, d50, d200, d7, d125, d720, d3, d80, d450, d17``
####  System B: ``d2, d300, d700, d55, d230, d999, d600, d80, d480, d33``

1. Compute and Plot Precision and Recall for each system:

    1. As a function of the number of documents returned (1 to 10).
    2. Show precision and recall values after each document is retrieved.

2. Plot the Precision-Recall curve for:
    1. System A
    2. System B

3. Calculate the F1-score for each system after retrieving the top 10 documents.

4. Compare the performance of both systems:
    1. Which system performs better?
    2. Justify your answer using precision, recall, and F1-score insights.
    3. Discuss any trade-offs observed between precision and recall. 


## Formulas:


$$
\text{Precision} = \frac{\text{Number of Relevant Documents Retrieved}}{\text{Total Number of Documents Retrieved}}
$$



$$
\text{Recall} = \frac{\text{Number of Relevant Documents Retrieved}}{\text{Total Number of Relevant Documents in Collection}}
$$




> F1-score balances **precision** and **recall** using the harmonic mean:

$$
F1 = \frac{2 \cdot P \cdot R}{P + R}
$$

## Step 2: Plot Graph

```python
import matplotlib.pyplot as plt

# Precision and recall values
system_a_precision = [1.0, 0.5, 0.667, 0.5, 0.6, 0.667, 0.571, 0.625, 0.667, 0.6]
system_a_recall    = [0.033, 0.033, 0.067, 0.067, 0.1, 0.133, 0.133, 0.167, 0.2, 0.2]

system_b_precision = [1.0, 1.0, 1.0, 0.75, 0.8, 0.667, 0.714, 0.75, 0.778, 0.7]
system_b_recall    = [0.033, 0.067, 0.1, 0.1, 0.133, 0.133, 0.167, 0.2, 0.233, 0.233]

ranks = list(range(1, 11))

# Plot System A
plt.figure(figsize=(7, 5))
plt.plot(system_a_recall, system_a_precision, marker='o', color="blue", label="System A")
for r, p, k in zip(system_a_recall, system_a_precision, ranks):
    plt.text(r, p, f"A{k}", fontsize=8, color="blue")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve: System A")
plt.ylim(0, 1.05)
plt.xlim(0, 0.25)
plt.grid(True)
plt.legend()
plt.show()

# Plot System B
plt.figure(figsize=(7, 5))
plt.plot(system_b_recall, system_b_precision, marker='s', color="green", label="System B")
for r, p, k in zip(system_b_recall, system_b_precision, ranks):
    plt.text(r, p, f"B{k}", fontsize=8, color="green")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve: System B")
plt.ylim(0, 1.05)
plt.xlim(0, 0.25)
plt.grid(True)
plt.legend()
plt.show()
```

## Step 3: F1-Score (after 10 docs)

$$
F1 = \frac{2 \cdot P \cdot R}{P + R}
$$

* **System A**:

  * Precision = 6/10 = 0.60
  * Recall = 6/30 = 0.20
  * $F1 = \frac{2 \cdot 0.6 \cdot 0.2}{0.6 + 0.2} = 0.30$

* **System B**:

  * Precision = 7/10 = 0.70
  * Recall = 7/30 = 0.233
  * $F1 = \frac{2 \cdot 0.7 \cdot 0.233}{0.7 + 0.233}  0.35$


## Step 4: Comparison

1. **Which performs better?**
    **System B** performs better overall.

2. **Justification:**

   * Higher **precision** (0.70 vs 0.60)
   * Higher **recall** (0.233 vs 0.20)
   * Higher **F1-score** (0.35 vs 0.30)

3. **Trade-offs observed:**

   * **System A** retrieves relevant docs steadily but is hurt by early non-relevant results.
   * **System B** front-loads relevant docs (first 3 all relevant)  higher early precision, smoother PR curve.
   * Both systems have **low recall** because they only return 10 docs out of 1500 (and only 30 are relevant).
