# Baseband Signal vs Carrier Signal

## Baseband Signal: 
The original signal, that contains actual information(data) before it is shifted to higher frequency for transmission.
- It is centered around 0Hz, meaning it has low frequency components.
- Baseband signal cannot be transmitted directly, but with the help of high frequency periodic signal, it can be transmitted.

## Carrier Signal:
The high frequency periodic signal, which carries baseband signal (original data) for far communication is called carrier signal.

# Modulation

Modulation is the process, in which one of the properties of the signal is changed according to baseband signal.

And based on these 3 properties of signal (amplitude, phase, frequency), there are different types of modulation.

# Why modulation is required. 

## **1. To reduce the antenna size.**

### Antenna:
- An antenna is a device that radiates or receives electromagnetic waves (radio signals).
- It converts electrical signals from your transmitter into electromagnetic waves in the air, and vice versa for reception.

### Antenna Size:

* **Antenna size** usually refers to the **physical length of the antenna**.
* The **length is related to the wavelength ($\lambda$)** of the signal it is designed to transmit or receive.
* A common design rule:

$$
\text{Antenna length} \approx \frac{\lambda}{2} \text{ or } \frac{\lambda}{4}
$$


### Why does wavelength matter?

* **Wavelength ($\lambda$)** depends on frequency:

$$
\lambda = \frac{c}{f}
$$

where $c = 3\times10^8$ m/s (speed of light) and $f$ = frequency in Hz.

* Lower frequency → longer wavelength → longer antenna.
* Higher frequency → shorter wavelength → shorter antenna.

**Example:**

1. Baseband digital signal: 1 kHz → wavelength = 300 km → impossible antenna (300/4 km)
2. FM radio: 100 MHz → wavelength = 3 m → antenna \~1.5 m (half-wave) → practical. (0.4 m)
### **Key idea**

* The **antenna size** must “match” the wavelength to efficiently transmit or receive the signal.
* If it’s too short or too long compared to the wavelength, **radiation efficiency drops**, and signal strength is weak.


## **2. To reduce the interference.**

- Many devices transmit at the same time. Interference may happen.
- Modulation allows different signals to occupy different frequency bands, avoiding interference.
- Example: Radio stations use different carrier frequencies.

<img src="img/Modulation9.jpg" width="800">

1. **PAM:** In case of pulse amplitude modulation, the width or duration of the pulse remains the same, but height or amplitude of the pulse signal wil change according to the message signal. 

<img src="img/Modulation10.jpg" width="800">

2. **PWM:** Similarly in case of pulse width modulation, the width or duration of the pulse remains the same, but width of the pulse signal will change according to the message signal. 

2. **PPM:** Similarly in case of pulse position modulation, the width or duration of the pulse remains the same, but position of the modulated signal changes wrt the carrier signal with the change in the message signal.

> **Note:** PPM signal, can also be generated using PWM signal.

4. **PCM:** In pulse code modulation, the message signal is sampled at the finite interval, and the sample value is quantized using the quantization process. And using the analogue to digital conveter, the message signal is encoded.

# PSK: Phase shift Keying

## Definition of PSK:
In PSK, carrier phase is changing accoring to the amplitude of the message signal. 

## Waveforms of PSK

In binary psk (there are 2 phases, 0° and 180°). And phase shift happends of 180 degree. 

- 1 -> 0°  phase
- 0 -> 180° phase

<img src="img/Modulation16.jpg" width="800">

## Multilevel PSK

1. In BPSK, we have L = 2 = 2^n -> n=1 -> 
    - ɸ = 0° (0)
    - ɸ = 180° (1)

2. In QPSK, we have L = 4 = 2^2n -> n=2 -> 
    - ɸ = 0° (00)
    - ɸ = 90° (01)
    - ɸ = 180° (10)
    - ɸ = 270° (11)

3. Also, we have L = 8 = 2^3n -> n=3 -> 
    - ɸ = 0°
    - ɸ = 45°
    - ɸ = 90°
    - ɸ = 135°
    - ɸ = 180°
    - ɸ = 225°
    - ɸ = 270°
    - ɸ = 315°
